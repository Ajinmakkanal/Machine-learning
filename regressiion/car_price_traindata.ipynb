```python
#for encoding the NAMe it is very long and we need to split it in such a way that we need to collect the brands only so that
# we use 

df['Brand']=df['Name'].apply(lambda x:x.split() [0])
df['Brand']
#we use apply otherwise it will not work in a dataset if any updation occurs
#lambda num1,num2:num1+num2
```

```python
#NOW ENCODING 
##3rd encoding

#GET_DUMMIES


#get_dummies is an encoding technique in pandas

#it works similar to one-hot encoding but in here the first column of the encoded columns is dropped

df6=pd.get_dummies(df[['Location','Fuel_Type','Transmission','Owner_Type','Brand']],drop_first=True,dtype=int)
df6                
```

```python
df6.shape
```

```python
df8=pd.concat([df,df6],axis=1)
df8
```
