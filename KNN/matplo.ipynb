```python
#marker properties

plt.plot(x,y,marker='o',ms=20,mfc='g',mec='r') 
#ms==marker size
#mfc=marker face color
# mec=marker edge color
```

```python
height=[150,160,155,180]
weight=[45,50,51,58]
plt.plot(height,weight)
plt.xlabel('height in cm',color='r',fontsize='10')
plt.ylabel('weight in kg',color='g',fontsize='10')
plt.title('Height_Weight Graph',color='b',fontsize='20')
```

```python
velocity=[10,20,30,40,50]
displacement=[5,10,15,20,25]
plt.plot(velocity,displacement)
plt.xlabel('velocity in m/s',color='b',fontsize='10')
plt.ylabel('displacement in m/s',color='b',fontsize='10')
plt.title('VELOCITY_DISPLACEMENT GRAPH',color='r',fontsize='15')
plt.grid(linestyle='dotted',color='g',linewidth='3') #providing a grid, axis=x provides line corresponding to x axis and vise versa
```

```python
#SCATTER PLOT
#it is used for representing points
x1=[1,6,8,5,6,4,5]
x2=[3,5,8,9,2,3,5]
plt.scatter(x1,x2,color='r')
```

```python
#SUBPLOT
#if we have to plot multiple graph in a single graph paer we would use subplot to partition it into 2 graph paper

x1=[1,2,3,4]
y1=[5,6,7,8]
plt.subplot(1,2,1)  #here 3rd argument 1,2,1 that is 1 decides which graph would be the first in the order
plt.plot(x1,y1)     #first and second argument decides the rows and columns
plt.title('x_graph')
x2=[3,4,6,1]
y2=[1,7,2,8]
plt.subplot(1,2,2)
plt.plot(x2,y2)
plt.title('y_graph')

plt.suptitle('GRAPH-1 GRAPH-2')  #IT IS USED FOR PROVIDING A MAIN HEADING  
```

```python
#BAR CHART
#WE NEED BOTH X AND Y AXIS
X=[5,8,9,3]
Y=[3,6,9,2]
plt.bar(X,Y)
```

```python
# PIE CHART
x1=[10,20,30,40]
var=['apple','orange','grape','mango']  #providing labels
colo=['black','red','yellow','blue']
plt.pie(x1,labels=var,colors=colo,autopct='%1.0f%%')
plt.legend(loc='lower right') #provide a descripiton of colors
plt.tight_layout() #reduciing overlapping
```

```python
import numpy as np
import pandas as pd
df=pd.read_csv('D:/customer1.txt',header=None)
df.columns=['id','fname','lname','age','prof','loc']
df
```

```python
import matplotlib.pyplot as plt
plt.bar(df1.index,df1.values) #(x,y)
plt.xlabel('location')
plt.ylabel('count')
plt.title('LOCATION COUNT DATA')
```
